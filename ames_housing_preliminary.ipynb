```python
# Is the training set representative of the test set in terms of missing values?

def find_missing(dataset, return_feature_names=False):
  
  from IPython.display import display
  
  missing_values = pd.DataFrame(dataset.isnull().sum())
  missing_percent = pd.DataFrame(missing_values/len(dataset)*100)
  missing = pd.concat([missing_values, missing_percent], axis=1)
  missing.columns = ['Missing values', '%']
  missing = missing.sort_values(by='Missing values', ascending=False)

  display(missing[missing['Missing values'] > 0])

  if return_feature_names == True:
    return missing[missing['Missing values'] > 0].index.values.tolist()
```

This tells us that the train set is not entirely representative of the test set. Based on information from `data_description.txt`, we need to do the following:
* **Account for missing values in numerical features:** If these features do not contain missing data in the train set, we can't do anything about it, since we are treating the test set as unseen data.

* **Look at all categorical features:** We need to look at all categorical features, not just those with 'missing' values, and figure out how to deal with it.

```python
class Preprocessor:

  def __init__(self, cat_cols_spec=None, cat_cols_reg=None, num_cols=None, ord_cols=None):
    self.cat_cols_spec = cat_cols_spec
    self.cat_cols_reg = cat_cols_reg
    self.num_cols = num_cols
    self.ord_cols = ord_cols
    self.mode_cat_cols_reg = pd.DataFrame()
    self.mode_ord_cols = pd.DataFrame()
    self.mean_num_cols = pd.DataFrame()

  def fit(self, X, y=None):
    '''
    Learn information from the training data to transform the test data. 
    Only call fit on train data.
    '''
    

    if self.cat_cols_reg:
      for col in self.cat_cols_reg:
        self.mode_cat_cols_reg[col] = X[col].mode()
    
    if self.num_cols:
      for col in self.num_cols:
        self.mean_num_cols[col] = round(X[col].mean())

    if self.ord_cols:
      for col in self.ord_cols:
        self.mode_ord_cols[col] = X[col].mode()

    return  self


  def transform(self, X, y=None):
    '''
    Use fit information if applicable and impute values to train and test data.
    '''
    X_new = X

    if self.cat_cols_spec:
      for col in self.cat_cols_spec:
        X_new[col] = X_new[col].fillna('None')
    
    if self.cat_cols_reg:
      for col in self.cat_cols_reg:
        X_new[col] = X_new[col].fillna(self.mode_cat_cols_reg[col])
    
    if self.num_cols:
      for col in self.num_cols:
        X_new[col] = X_new[col].fillna(self.mean_num_cols[col])

    if self.ord_cols:
      for col in self.ord_cols:
        X_new[col] = X_new[col].fillna(self.mode_ord_cols[col])

    X_new = X_new.dropna()

    return X_new


  def fit_transform(self, X, y=None):
    '''
    Combines fit and transform for use in sklearn pipelines
    '''
    return self.fit(X).transform(X)
```

```python
prep = Preprocessor(
    categorical_features_special,
    categorical_features_regular,
    numeric_features,
    ordinal_features
)
```
